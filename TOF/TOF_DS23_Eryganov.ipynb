## Описание проекта

### Цель проекта

Цель данного проекта - построить модель, которая будет давать рекомендации типа операции при заболевании сердца "Тетрада Фалло".

### Постановка задачи

Тетрада Фалло — так называемый синий порок сердца, сочетающий, по определению французского врача Этьена-Луи Артура Фалло, четыре аномалии:
* стеноз выходного отдела правого желудочка (клапанный, подклапанный, стеноз лёгочного ствола и (или) ветвей лёгочной артерии либо комбинированный);
* высокий (субаортальный) дефект межжелудочковой перегородки;
* декстрапозиция аорты;
* гипертрофия правого желудочка (как следствие затруднённого оттока крови из желудочка).

В проекте необходимо построить модель, которая будет давать рекомендации конкретного типа операции для ребенка из 4 типов операций:
1. радикальная, 
2. паллиативная 1 с искусственным кровообращением (ИК), 
3. паллиативная 2 без ИК,
4. паллиативная миниинвазиваная без ИК.

### Условия заказчика

- Train/test split 70/30, stratified,
- Random state  - 654321,
- Метрика - F1 Macro,
- Обработать пропуски, возможно там нули,
- Метрики на train и test не должны сильно отличаться.### Условия заказчика.

### Описание данных

Данные представлены в двух таблицах:
* `TOF_name_type_procedure.xlsx` - таблица с названиями признаков.
* `Base_TOF.xlsx` - таблица с данными для обучения модели - медицинскими показателями, связанными с патологией. Тетрада Фалло - порок, который очень часто требует несколько операций. Датасет содержит данные о первых операциях, которые были сделаны детям.

Признаки:

* `пол` — Пол пациента (мужской или женский);
* `возраст` — Возраст пациента на момент обследования или операции, дней;
* `вес` — Масса тела пациента, кг;
* `рост` — Рост пациента, см;
* `бса(Body Surface Area, BSA)` — Площадь поверхности тела, м2. Обычно рассчитывается на основе роста и веса пациента и используется для различных медицинских целей, таких как дозировка препаратов;
* `сопвпс (Сопутствующие заболевания)` — Информация о наличии других заболеваний, которые могут влиять на состояние пациента или лечение основного заболевания;
* `гествозраст (Гестационный возраст)` — Длительность беременности на момент обследования, если применимо (например, для новорожденных), дней;
* `О2капиляр (Капиллярное насыщение кислородом)` — Уровень насыщения крови кислородом, измеряемый в капиллярах, %;
* `одышцианприст (Одышка и цианоз при физической нагрузке)` — Наличие одышки и цианоза (синюшность кожи) при физической активности;
* `фвлжпервич (Фракция выброса левого желудочка на первичном обследовании)` — Показатель, отражающий эффективность работы левого желудочка сердца, %;
* `кдрлж (Конечный диастолический размер левого желудочка)` — Размер левого желудочка сердца в конце диастолы (расслабления), , мм;
* `ксрлж (Конечный систолический размер левого желудочка)` — Размер левого желудочка сердца в конце систолы (сокращения), мм;
* `кдолж (Конечный диастолический объем левого желудочка)` — Объем крови в левом желудочке сердца в конце диастолы, мл;
* `устьеПЛА (Устье правой легочной артерии)` — Диаметр устья (отверстия) правой легочной артерии, мм;
* `УстьеЛЛА (Устье левой легочной артерии)` — Диаметр устья левой легочной артерии, мм;
* `стволла (Ствол легочной артерии)` — Диаметр ствола легочной артерии, мм;
* `фкла (Фракция кровотока легочной артерии)` — Объемный процент крови, проходящей через легочную артерию;
* `створкиклЛА (Створки клапана легочной артерии)` — Оценка состояния створок клапана легочной артерии;
* `градиентдавленияпжла (Градиент давления между правым желудочком и легочной артерией)` — Разница в давлении между правым желудочком и легочной артерией, мм рт.ст.;
* `дмжпмм (Диаметр межжелудочковой перегородки в миллиметрах)` — Толщина межжелудочковой перегородки, мм;
* `давлениеПЖ (Давление в правом желудочке)` — Давление в правом желудочке сердца, мм рт.ст.;
* `ЛАИ (Легочный артериальный индекс)` — Индекс, учитывающий параметры легочной артерии для оценки легочной гипертензии;
* `ABнАо (Насыщение артериальной крови кислородом)` — Уровень насыщения крови кислородом в артериях, %;
* `операция` — Тип операции, проведенной пациенту - целевой признак.

````python
def df_research_method(df, method):
    """
    Применяет определенный метод изучения данных в датасете.

    Параметры:
    - df: Исследуемый датасет.
    - method: Название метода.
   
    Пример:
    ```
    df_research_method(df, 'head')
    ```
    """
    
    if method == 'head':
        display(df.head())
    elif method == 'sample':
        display(df.sample(10 ))
    elif method == 'shape':
        display(df.shape)
    elif method == 'describe':
        display(df.describe(include='all').T)
    elif method == 'duplicated':
        dp = df.duplicated().sum()
        display(dp)
        if dp > 0:
            print('Количество явных дубликатов:')
            display(df[df.duplicated()])
    elif method == 'unique':
        # for column in df.select_dtypes(include='object').columns:
        print('Количество уникальных значений в колонке')
        for column in df.columns:
            df_true = df[column].notnull()
            df_all = df.loc[df_true, column].count()
            df_temp = df.loc[df_true, column].sort_values().unique()
            if df_all != len(df_temp):
                print(f'{column}: {len(df_temp)} всего {df_all}')
                print(f'{column}: {df_temp}')
    elif method == 'msno':
        display(msno.bar(df))
        plt.show()
    else:
        display(df.info())

# Функция применения методов к списку датасетов
def df_research(df_list, df_methods):
    """
    Применяет метод из списка для изучения данных в списке датасетов.

    Параметры:
    - df_list: Список исследуемых датасетов (имя переменной, описание, имя файла).
    - df_method: Список названий методов.
   
    Пример:
    ```
    df_research(df_list, df_methods)
    ```
    """

    for method in df_methods:
        print(f'Изучим {method[1]}:')
        print('-' * 70)
        for df in df_list:
            print('Dataset:', df[1])
            print('-' * 70)            
            globals()[df[0]] = globals()[df[0]].applymap(lambda x: x.lower() if isinstance(x, str) else x)
            df_research_method(globals()[df[0]], method[0])
            print('-' * 70);
````

````python
def show_barplot(title, x, y, x_label, y_label):
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=x, y=y)
    plt.suptitle(title, fontsize=16, ha='center')
    ax.figsize=(16, 6) 
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.bar_label(ax.containers[0],
             color='black',
             fontsize=12,
             fontweight='bold')
    plt.show();
````

````python
# создадим функцию для построения диаграммы корреляции признаков
def corr_matrix(data, columns, interval_cols):
    """
    Отображает матрицу кореляции.

    Параметры:
    - data: Датасет.
    - columns: Названия колонок.
    - interval_cols: Числовые колонки.
    
   
    Пример:
    ```
    corr_matrix(data, columns, interval_cols)
    ```
    """
    
    # сформируем матрицу корреляции
    cor_matrix = data[columns].phik_matrix(interval_cols=interval_cols)
    # построим тепловую карту
    plt.figure(figsize=(18, 12))
    sns.heatmap(cor_matrix, 
                annot=True, 
                linewidths=0.5, 
                cmap='coolwarm')
    plt.show()
    return cor_matrix;
````

````python
# Подготовка функции создания моделей
def create_model(model_name, pipeline, param_grid, features_train, target_train, scorer):
    result = {}
    # Создание модели
    grid_search = GridSearchCV(pipeline, param_grid, n_jobs=-1, cv=5, scoring=scorer)
    # Обучение модели
    start_time = time.time()
    grid_search.fit(features_train, target_train)
    result['training_time'] = round(time.time() - start_time, 2)
    # Сохраненение модели с лучшими результами
    result['best_model'] = grid_search.best_estimator_
    # Формирование предсказаний
    start_time = time.time()
    result['predictions'] = result['best_model'].predict(features_train)
    if model_name != 'PassiveAggressiveClassifier':
        result['probas'] = result['best_model'].predict_proba(features_train)[:,1]
    else:
        result['probas'] = None
    result['prediction_time'] = round(time.time() - start_time, 2)
    # Сохранение лучшей метрики
    result[scorer] = round(grid_search.best_score_, 2)
    result['grid_search'] = grid_search
    print(f'Модель {model_name}')
    print('-' * 70)
    print(f'Время обучения: {result["training_time"]} сек.')
    print(f'Время предсказания: {result["prediction_time"]} сек.')
    print(f'{scorer.upper()}: {result[scorer]}')
    print(f'Лучшая модель:{result["best_model"]}')
    print('-' * 70)    
    return result;
````

````python
def show_metrics(predictions, y_test):
    f1 = f1_score(y_test, predictions, average='macro')
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions, average='macro')
    recall = recall_score(y_test, predictions, average='macro')
    print('F1: ', round(f1, 3))
    print('Accuracy: ', round(accuracy, 3))
    print('Precision: ', round(precision, 3))
    print('Recall: ', round(recall, 3))
    print('-' * 70)
    print('Матрица ошибок')
    cm = confusion_matrix(y_test, predictions)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.show();
````
