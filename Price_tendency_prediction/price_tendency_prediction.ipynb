## Описание проекта

### Цель проекта

Цель данного проекта - создать модель МО, которая будет определять направление изменения цены.

### Описание данных

Данные находятся в файле `test_ds.csv`. 

Признаки:
* `date` - дата,
* `open` - цена открытия, 
* `high` - цена максимум, 
* `low` - цена минимум, 
* `close` - цена закрытия,
* `vol` - объем, 
* `open_high` - разница открытия и максимум, 
* `open_low` - разница открытия и минимум, 
* `high_low` - разница максимум и минимум, 
* `high_close` - разница максимум и закрытия, 
* `low_close` - разница минимум и закрытия, 
* `close_open` - разница закрытия и открытия, 
* `percent` - процент изменения цены относительно предыдущего дня, 
* `tendency` - направление (1 - рост, 0 - падение), 
* `target` - целевая переменная (направление движения цены на следующий день).

````python
def df_research_method(df, method):
    """
    Применяет определенный метод изучения данных в датасете.

    Параметры:
    - df: Исследуемый датасет.
    - method: Название метода.
   
    Пример:
    ```
    df_research_method(df, 'head')
    ```
    """
    
    if method == 'head':
        display(df.head())
    elif method == 'sample':
        display(df.sample(10 ))
    elif method == 'shape':
        display(df.shape)
    elif method == 'describe':
        display(df.describe(include='all').T)
    elif method == 'duplicated':
        dp = df.duplicated().sum()
        display(dp)
        if dp > 0:
            print('Количество явных дубликатов:')
            display(df[df.duplicated()])
    elif method == 'unique':
        # for column in df.select_dtypes(include='object').columns:
        for column in df.columns:
            df_true = df[column].notnull()
            df_all = df.loc[df_true, column].count()
            
            df_temp = df.loc[df_true, column].sort_values().unique()
            if df_all != len(df_temp):
                print(f'Количество уникальных значений в колонке {column}: {len(df_temp)} всего {df_all}')
                # print(f'Уникальные значения в колонке {column}: {df_temp}')
    elif method == 'msno':
        display(msno.bar(df))
        plt.show()
    else:
        display(df.info())

# Функция применения методов к списку датасетов
def df_research(df_list, df_methods):
    """
    Применяет метод из списка для изучения данных в списке датасетов.

    Параметры:
    - df_list: Список исследуемых датасетов (имя переменной, описание, имя файла).
    - df_method: Список названий методов.
   
    Пример:
    ```
    df_research(df_list, df_methods)
    ```
    """

    for method in df_methods:
        print(f'Изучим {method[1]}:')
        print('-' * 70)
        for df in df_list:
            print('Dataset:', df[1])
            print('-' * 70)            
            globals()[df[0]] = globals()[df[0]].applymap(lambda x: x.lower() if isinstance(x, str) else x)
            df_research_method(globals()[df[0]], method[0])
            print('-' * 70);
````

````python
# создадим функцию для построения диаграммы корреляции признаков
def corr_matrix(data, columns, interval_cols):
    """
    Отображает матрицу кореляции.

    Параметры:
    - data: Датасет.
    - columns: Названия колонок.
    - interval_cols: Числовые колонки.
    
   
    Пример:
    ```
    corr_matrix(data, columns, interval_cols)
    ```
    """
    
    # сформируем матрицу корреляции
    corr_matrix = data[columns].phik_matrix(interval_cols=interval_cols)
    # построим тепловую карту
    plt.figure(figsize=(18, 12))
    sns.heatmap(corr_matrix, 
                annot=True, 
                linewidths=0.5, 
                cmap='coolwarm')
    plt.show();
````

````python
# Подготовка функции создания моделей
def create_model(model_name, pipeline, param_grid, features_train, target_train, scorer):
    result = {}
    # Создание модели
    grid_search = GridSearchCV(pipeline, param_grid, n_jobs=-1, cv=5, scoring=scorer)
    # Обучение модели
    start_time = time.time()
    grid_search.fit(features_train, target_train)
    result['training_time'] = round(time.time() - start_time, 2)
    # Сохраненение модели с лучшими результами
    result['best_model'] = grid_search.best_estimator_
    # Формирование предсказаний
    start_time = time.time()
    result['predictions'] = result['best_model'].predict(features_train)
    if model_name != 'PassiveAggressiveClassifier':
        result['probas'] = result['best_model'].predict_proba(features_train)[:,1]
    else:
        result['probas'] = None
    result['prediction_time'] = round(time.time() - start_time, 2)
    # Сохранение лучшей метрики
    result[scorer] = round(grid_search.best_score_, 2)
    result['grid_search'] = grid_search
    print(f'Модель {model_name}')
    print('-' * 70)
    print(f'Время обучения: {result["training_time"]} сек.')
    print(f'Время предсказания: {result["prediction_time"]} сек.')
    print(f'{scorer.upper()}: {result[scorer]}')
    print(f'Лучшая модель:{result["best_model"]}')
    print('-' * 70)    
    return result;
````

````python
def show_metrics(predictions, y_test):
    f1 = f1_score(y_test, predictions)
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)
    print('F1: ', round(f1, 3))
    print('Accuracy: ', round(accuracy, 3))
    print('Precision: ', round(precision, 3))
    print('Recall: ', round(recall, 3))
    print('-' * 70)
    print('Матрица ошибок')
    cm = confusion_matrix(y_test, predictions)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.show();
````

Колонки типа `int64`, `float64` заменим на `int32`, `float32` с целью экономии памяти.
Проверим результат.

````python
for df in datasets:
    print('Dataset:', df[1])
    # Приведем названия колонок к стандарту "snake_case".
    globals()[df[0]].columns = [x.lower().
                 split('(', 1)[0].split(',', 1)[0].replace(' ', '_').rstrip('_') for x in globals()[df[0]].columns]
    # Колонки типа `int64`, `float64` заменим на `int32`, `float32` с целью экономии памяти.
    for column in globals()[df[0]].columns:
        if globals()[df[0]][column].dtypes == 'int64':
            globals()[df[0]][column] = globals()[df[0]][column].astype('int32')
        if globals()[df[0]][column].dtypes == 'float64':
            globals()[df[0]][column] = globals()[df[0]][column].astype('float32')
    # Проверим результат.
    display(globals()[df[0]].info());
````

### Промежуточные выводы

В ходе предобработки данных:

- Колонки типа `int64`, `float64` заменили на `int32`, `float32` с целью экономии памяти.
- Привели названия колонок к `snake_case`.

````python
train, test = train_test_split(test_ds_new, shuffle=False, test_size=0.2)

features_train = train.drop('target', axis = 1)
target_train = train['target']
features_test = test.drop('target', axis = 1)
target_test = test['target'];
````

### Тестирование моделей

#### Логистическая регрессия `LogisticRegression`

````python
%%time
model_name = 'LogisticRegression'
model = LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', class_weight='balanced')

pipeline = Pipeline(steps = [
     ('models', model)
])
param_grid = {
    'models': [model],
    'models__C': range(1, 10)
}

models[model_name] = create_model(model_name,
                                  pipeline,
                                  param_grid,
                                  features_train,
                                  target_train,
                                  scorer);
````

#### Дерево решений `DecisionTreeClassifier`

````python
%%time
model_name = 'DecisionTreeClassifier'
model = DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced')


pipeline = Pipeline(steps = [
     ('models', model)
])

param_grid = {
    'models': [model],
    'models__max_depth': range(2, 5),
    'models__max_features': range(2, 5)
}

models[model_name] = create_model(model_name,
                                  pipeline,
                                  param_grid,
                                  features_train,
                                  target_train,
                                  scorer);
````

#### Случайный лес `RandomForestClassifier`

````python
%%time
model_name = 'RandomForestClassifier'
model = RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced')

pipeline = Pipeline(steps = [
     ('models', model)
])
param_grid = {
    'models': [model],
    'models__max_depth': range(2, 5),
    'models__n_estimators': range(5, 20, 5)
}

models[model_name] = create_model(model_name,
                                  pipeline,
                                  param_grid,
                                  features_train,
                                  target_train,
                                  scorer);
````

#### Пассивно-агрессивный алгоритм `PassiveAggressiveClassifier`

````python
%%time
model_name = 'PassiveAggressiveClassifier'
model = PassiveAggressiveClassifier(max_iter=200, class_weight='balanced')

pipeline = Pipeline(steps = [
     ('models', model)
])
param_grid = {
    'models': [model]
}

models[model_name] = create_model(model_name,
                                  pipeline,
                                  param_grid,
                                  features_train,
                                  target_train,
                                  scorer);
````
